# 5 Создаем таблицу user_info

```python
# Преобразование времени в формат datetime
oplogs['user_time'] = pd.to_datetime(oplogs['user_time'])

# Расчитываем недостающие значения
def extract_unique_dates(series):
    return list(series.dt.date.unique())
    
def determine_payer(values):
    return 1 if 'purchase' in values.values else 0

def calculate_total_revenue(group):
    return group.loc[group['event_name'] == 'purchase', 'param_2'].sum()
    
def extract_last_value(series):
    return series.iloc[-1] if not series.empty else 0
    
def extract_bronze_cup(extra_2_series):
    return (extra_2_series == 'bronze_cup').sum()

def extract_silver_cup(extra_2_series):
    return (extra_2_series == 'silver_cup').sum()

def extract_gold_cup(extra_2_series):
    return (extra_2_series == 'gold_cup').sum()
    
# Агрегация данных для новой таблицы
user_info = oplogs.groupby('user_id').agg(
    platform_type = ('platform_type', 'first'),
    country = ('country', 'first'),
    first_version = ('version', 'min'),
    last_version = ('version', 'max'),
    type_trafic = ('extra_1', 'first'),
    first_time = ('user_time', 'min'),
    last_time = ('user_time', 'max'),
    dates = ('user_time', extract_unique_dates),
    payer = ('event_name', determine_payer),
    total_revenue = ('param_2', lambda x: calculate_total_revenue(oplogs.loc[x.index])),
    max_level = ('level', 'max'),
    amount_gold = ('param_2', extract_last_value),
    amount_bronze_cup = ('extra_2', extract_bronze_cup),
    amount_silver_cup = ('extra_2', extract_silver_cup),
    amount_gold_cup = ('extra_2', extract_gold_cup)
).reset_index()
user_info
```

# 6 Создаем таблицу user_session

```python
# Преобразование user_time в формат datetime
oplogs['user_time'] = pd.to_datetime(oplogs['user_time'])

# Добавление вспомогательных колонок для расчёта
oplogs['is_bronze_cup'] = oplogs['extra_1'] == 'bronze_loc'
oplogs['is_silver_cup'] = oplogs['extra_1'] == 'silver_loc'
oplogs['is_gold_cup'] = oplogs['extra_1'] == 'gold_loc'
oplogs['amount_get_level'] = oplogs['is_bronze_cup'] + oplogs['is_silver_cup'] +  oplogs['is_gold_cup']

# Группировка данных и расчёты
user_session = oplogs.groupby(['user_id', 'session_id']).agg(
    time_start_session = ('user_time', 'min'),
    time_finish_session = ('user_time', 'max'),
    length_session = ('user_time', lambda x: (x.max() - x.min()).total_seconds()),
    amount_get_level = ('amount_get_level', 'sum'),
    amount_get_gold = ('param_1', extract_last_value),
    amount_get_bronze_cup = ('is_bronze_cup', 'sum'),
    amount_get_silver_cup = ('is_silver_cup', 'sum'),
    amount_get_gold_cup = ('is_gold_cup', 'sum')
).reset_index()
user_session
```

# 12 Создадим таблицу с данными об A/B-эксперименте

```python
# Преобразование столбца 'user_time' в формат datetime
oplogs['user_time'] = pd.to_datetime(oplogs['user_time'])

# Фильтрация по диапазону дат
df_filtered = oplogs[(oplogs['user_time'] >= '2023-02-03') & (oplogs['user_time'] <= '2023-02-11')]

# Фильтрация по значениям event_name
df_filtered = df_filtered[df_filtered['event_name'].isin(['stage_start', 'stage_win'])]

# Фильтрация по значению param_1 = 2
df_filtered = df_filtered[df_filtered['param_1'] == 2]

# Фильтрация по значению param_2 = 1
df_test = df_filtered[df_filtered['param_2'] == 1]
df_test
```

# 13 Выведем результат A/B-эксперимента

```python
# Добавление группы на основе user_id
df_test['group'] = df_test['user_id'] % 10
df_test['group'] = df_test['group'].apply(lambda x: 'A' if x in [0, 1, 2, 3, 4] else 'B')

# Подсчет событий stage_start и stage_win для каждой группы
results = df_test.groupby(['group', 'event_name']).size().unstack(fill_value=0)

# Вычисление уровня завершения (completion rate)
results['completion_rate'] = results['stage_win'] / results['stage_start']
results
```

# 14 Проверяем статистическую значимость эксперимента

```python
from statsmodels.stats.proportion import proportions_ztest

# Количество начавших уровень
n_A = results.loc['A', 'stage_start']
n_B = results.loc['B', 'stage_start']

# Количество завершивших уровень
x_A = results.loc['A', 'stage_win']
x_B = results.loc['B', 'stage_win']

# Выполнение z-теста
counts_win = [x_A, x_B]
counts_start = [n_A, n_B]
z_stat, p_value = proportions_ztest(counts_win, counts_start)

print(f"Z-Statistic: {z_stat}")
print(f"P-Value: {p_value}")

# Проверка статистической значимости
alpha = 0.05
if p_value < alpha:
    print("Разница между группами статистически значима")
else:
    print("Разница между группами статистически НЕ значима")
```
